### Visualization of examples from Tensorboard
Small illustraion of what some traininig PSNR and SSIM scores look like from tensorboard

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import ticker
import matplotlib.ticker as mticker
```

```python
iters_per_epoch = 14738
```

```python
@ticker.FuncFormatter
def human_format(num, pos):
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    # add more suffixes if you need them
    return '%.0f%s' % (num, ['', 'K', 'M', 'G', 'T', 'P'][magnitude])
```

```python
def path_to_xy(path, iters_per_epoch = 14738):
    df = pd.read_csv(path)
    x = df['Step'].to_numpy() * iters_per_epoch
    y = df['Value'].to_numpy()
    return x,y
```

```python
def create_plot(paths, names, ylabel, figname=None):
    fig, ax = plt.subplots(figsize = (7,7), dpi = 1200)
    ax.xaxis.set_major_locator(mticker.MultipleLocator(base=100000))
    if ylabel == "PSNR":
        ax.yaxis.set_major_locator(mticker.MultipleLocator(base=0.5))
    else:
        ax.yaxis.set_major_locator(mticker.MultipleLocator(base=0.1))


    ax.grid()
    # plt.gca().set_aspect(iters_per_epoch * 5)

    ax.set_xlim(iters_per_epoch, 40*iters_per_epoch)
    if ylabel == "PSNR":
        ax.set_ylim(24,32)#(np.min(y),np.max(y)+0.25)
    else:
        ax.set_ylim(0.3,1)

    ax.xaxis.set_major_formatter(human_format)
    for spine in ax.spines.values():
        spine.set_edgecolor('grey')

    # ax.set_aspect(ax.get_xlim()[1]/ax.get_ylim()[1] * 10) 
    plt.tick_params(top=False, bottom=False, left=False, right=False)
    
    ax.set_xlabel("Iteration")
    ax.set_ylabel(ylabel)

    for name, (x,y) in zip(names, [path_to_xy(path) for path in paths]):
        ax.plot(x, y, label=name)

    ax.legend(loc='lower right', prop={'size': 10})
    if figname:
        plt.savefig(f"tensorboard/{figname}", bbox_inches='tight')
    plt.show()
    
```

## Create plots for ST and Patchwise-ST for VGG- and Discriminator losses
### PSNR

```python
paths = [
    'data/tensorboard/st-patchwise-st-c-cd/psnr/ablation-patchwise-st.csv',
    'data/tensorboard/st-patchwise-st-c-cd/psnr/patchwise-st-cd-no-pixel.csv',
    'data/tensorboard/st-patchwise-st-c-cd/psnr/st-c-no-pixel-reweighted.csv',
    'data/tensorboard/st-patchwise-st-c-cd/psnr/st-cd-no-pixel-reweighted.csv'
]
names = [
    "Patchwise-ST w. VGG content loss",
    "Patchwise-ST w. Discriminator content loss",
    "ST w. VGG content loss",
    "ST w. Discriminator content loss"
]
create_plot(paths, names, "PSNR", "st-patchwise-st-c-cd-psnr.svg")
```

#### ssim

```python
paths = [
    'data/tensorboard/st-patchwise-st-c-cd/ssim/patchwise-st-cd-no-pixel(1).csv',
    'data/tensorboard/st-patchwise-st-c-cd/ssim/ablation-patchwise-st(1).csv',
    'data/tensorboard/st-patchwise-st-c-cd/ssim/st-cd-no-pixel-reweighted(1).csv',
    'data/tensorboard/st-patchwise-st-c-cd/ssim/st-c-no-pixel-reweighted(1).csv'
]
names = [
    "Patchwise-ST w. VGG content loss",
    "Patchwise-ST w. Discriminator content loss",
    "ST w. VGG content loss",
    "ST w. Discriminator content loss"
]
create_plot(paths, names, "SSIM", "st-patchwise-st-c-cd-ssim.svg")
```
